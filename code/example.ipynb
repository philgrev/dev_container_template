```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# define a function with type hints
def an_example_function(a: np.ndarray):
    """An example function that normalizes a numpy array."""
    return (a - np.min(a)) / (np.max(a) - np.min(a))
```
